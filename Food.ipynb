# Data Analysis on Indian foods

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from collections import Counter
```

```python
#Reading the CSV file and displaying it for viewing the data

food = pd.read_csv("indian_food.csv")
food
```

# Analysing the most commonly used ingredients in South India

```python
#Removing the ',' values in ingredients for easier filtering

food['ingredients'] = food['ingredients'].str.replace('[\,]', '')
food['ingredients'] = food['ingredients'].str.title()
# Filtering and obtaining data from South region

south = food.loc[food['region'] == 'South']
south.head()
```

```python
#Finding the top 10 commonly used ingredients and their counts

common_ing = Counter(" ".join(south["ingredients"]).split()).most_common(10)
common_ing
```

```python
#Placing the ingredients and their counts in seperate variables for easier plotting

items = [val[0] for val in common_ing]
values = [val[1] for val in common_ing]
```

```python
#Plotting the data

fig, ax = plt.subplots(figsize = (10,7))
ax.pie(values,labels = items, autopct='%1.2f%%')
ax.set(title = "Most used ingredients in South India");
```
